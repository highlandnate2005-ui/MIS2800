```python
#Block 1
# Instructions: Create a table named stock_performance with columns for ticker, price_before_game, price_week_after_game, trading_volume_millions
%%sql
CREATE TABLE IF NOT EXISTS stock_performance (
    ticker TEXT,
    price_before_game REAL,
    price_week_after_game REAL,
    trading_volume_millions INTEGER
) STRICT;
```

```python
# Block 2
# Instructions: Add the below values into the stock_performance table
%%sql
INSERT INTO stock_performance (ticker, price_before_game, price_week_after_game, trading_volume_millions)
VALUES
  ('PEP', 168.50, 172.10, 45),
  ('GOOGL', 142.80, 145.20, 110),
  ('AMZN', 170.20, 178.45, 95),
  ('VZ', 40.10, 39.85, 60),
  ('TSLA', 190.00, 185.20, 150),
  ('TMUS', 197.68, 199.34, 100),
  ('STLA', 22.40, 23.10, 15);
```

```python
#Block 3
# Instructions: Create a table named Ad_Buys with columns for ticker, company_name, ad_cost_millions, airtime_seconds.
# The ticker should be a TEXT data type and serve as the Primary Key.
# Select an appropriate data_type for the other columns.
# Remember to use the STRICT keyword to ensure the datatype is enforced.
%%sql
CREATE TABLE IF NOT EXISTS Ad_Buys (
    ticker TEXT PRIMARY KEY,
    company_name TEXT,
    ad_cost_millions REAL,
    airtime_seconds INTEGER
) STRICT;
```

```python
#Block 5
# Instructions: Show all the columns from the stock_performance table
%%sql
SELECT * FROM stock_performance;
```

```python
#Block 6
# # Instructions: Show the ticker and price before the game
# sort by ticker in alphabetical order from the stock_performance table
%%sql
SELECT ticker, price_before_game FROM stock_performance ORDER BY ticker ASC;
```

```python
#Block 7
# # Instructions: Show all the columns for the brands that have a ticker
# beginning with the letter T in the stock_performance table
%%sql
SELECT * FROM stock_performance WHERE ticker LIKE 'T%';
```

```python
#Block 8
# # Instructions: Show only the ticker and price_after_game
# from the stock_performance table if the price is above 150
%%sql
SELECT ticker, price_week_after_game FROM stock_performance WHERE price_week_after_game > 150;
```

```python
#Block 9
# Instructions: Show all the columns from the ad_buys table
%%sql
SELECT * FROM Ad_Buys;
```

```python
#Block 10
# # Instructions: Show the ticker and company_name from the ad_buys table in
# sort in descending order by company_name
%%sql
SELECT ticker, company_name FROM Ad_Buys ORDER BY company_name DESC;
```

```python
#Block 11
# # Instructions: Show the ticker and airtime seconds from the ad_buys table
# Filter to show only if the ticker begins with T and and airtime seconds is 0
%%sql
SELECT ticker, airtime_seconds FROM Ad_Buys WHERE ticker LIKE 'T%' AND airtime_seconds = 0;
```
