## YOUR ANSWERS TO PART 1 BELOW

**Question 1:**                                                              
Shifting from a simple data storage to Redshift data warehouse allowed Lyft's data scientists to analyze more histroical data instead of just current transactions. With Redshift, their data team could run simulations to identify ride overlap patterns. This analysis revealed that many rides had similar routes at peak times, which led to the creation of Lyft Line.


**Question 2:**                                                             
Lyft cannot use the same database for both OLTP and OLAP because they do different jobs. OTLP databases are designed for real-time tasks, such as booking rides and updating locations. Data warehouses are built for analyzing large amounts of past data. If Lyft ran big analysis queries on the same system that runs the app, it would slow everything down for users. Keeping them separate enables the app to remain fast while still providing valuable insights.

## YOUR ANSWERS TO PART 2 BELOW

**Question 1:**                                                              
Columnar storage means Amazon Redshift stores data by column instead of by row. This makes queries much faster because most analytical queries only need a few columns, not the entire row. Instead of reading all the data in a table, Redshift only reads the columns required, which reduces disk usage and speeds up performance. It also allows better data compression since each column stores similar types of data.

**Question 2:**                                                             
The Leader Node and Compute Node architecture makes BI reports faster because the work is divided and processed at the same time. The Leader Node receives the query and creates a plan, then sends parts of the work to multiple Compute Nodes. Each compute node processes its portion of the data in parallel, and then the leader combines the results and returns the final report. Because many nodes are working at once instead of one system doing everything, reports are generated much more quickly.
