# Task #1: Create and Insert (C)


```python
# create a TABLE titled products
%%sql
CREATE TABLE products (
  product_id INTEGER PRIMARY KEY,
  product_name TEXT,
  price REAL
) STRICT;
```

```python
# INSERT NSERT INTO products (name, price) VALUES ('Superbowl Jersey', 120.00)
%%sql
INSERT INTO products (product_id, product_name, price)
VALUES
  (1,'Superbowl Jersey', 120.00),
  (2, 'NFL Football', 95.00),
  (3, 'NFL Cleats', 100.00),
  (4, 'Ref Whistle', 15.00);
```

# Task #2: Read and Filter (R)

```python
# Show all products that cost more than $50
%%sql
SELECT * FROM products WHERE price > 50;
```

# Task 3: Update and Delete (U & D)

```python
# Update the price from 'Superbowl Jersey' to 99.99 and delete product_id 1
%%sql
UPDATE products SET price = 99.99 WHERE product_name = 'Superbowl Jersey';
Select * FROM products;
```

```python
# Delete product_id 1
%%sql
DELETE FROM products WHERE product_id = 1;
Select * FROM products;
```
